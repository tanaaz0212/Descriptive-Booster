```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_excel("C:\\Users\\tanaa\\Downloads\\household_dataset.xlsx")

# Basic Info
print(df.info())

# Central Tendency
print("Mean:", df["Household_Income"].mean())
print("Median:", df["Household_Income"].median())
print("Mode:", df["Household_Income"].mode()[0])

# Dispersion
print("Range:", df["Household_Income"].max() - df["Household_Income"].min())
print("Variance:", df["Household_Income"].var())
print("Std Dev:", df["Household_Income"].std())

# Skewness & Kurtosis
print("Skewness:", df["Household_Income"].skew())
print("Kurtosis:", df["Household_Income"].kurt())

# Histogram
plt.figure()
sns.histplot(df["Household_Income"], kde=True)
plt.title("Income Distribution")
plt.show()

# Boxplot
plt.figure()
sns.boxplot(x="Education_Level", y="Household_Income", data=df)
plt.title("Income vs Education")
plt.show()

# Age vs Income
plt.figure()
sns.scatterplot(x="Age_of_Household_Head", y="Household_Income", data=df)
plt.title("Age vs Income")
plt.show()
```
